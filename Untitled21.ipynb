```python
import pandas as pd
import numpy as np
import torch
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, f1_score, classification_report
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments
)
from sentence_transformers import SentenceTransformer, util
```

```python
train_df = pd.read_parquet(r"/content/drive/MyDrive/ILSI/train-00000-of-00001.parquet")
dev_df = pd.read_parquet(r"/content/drive/MyDrive/ILSI/dev-00000-of-00001 (1).parquet")
test_df = pd.read_parquet(r"/content/drive/MyDrive/ILSI/test-00000-of-00001.parquet")
statutes = pd.read_parquet(r"/content/drive/MyDrive/ILSI/statutes-00000-of-00001.parquet")

print("Train:", train_df.shape)
print("Dev:", dev_df.shape)
print("Test:", test_df.shape)
print("Statutes:", statutes.shape)
```

```python
# Convert list or array labels to plain strings
for df in [train_df, dev_df, test_df]:
    df["labels"] = df["labels"].apply(
        lambda x: x[0] if isinstance(x, (list, np.ndarray)) and len(x) > 0 else str(x)
    )

# Label encoding for statute IDs (like Section 2, Section 3)
le = LabelEncoder()
train_df["label_id"] = le.fit_transform(train_df["labels"])
dev_df["label_id"] = le.transform(dev_df["labels"])
test_df["label_id"] = le.transform(test_df["labels"])

num_labels = len(le.classes_)
print("Number of unique statute sections:", num_labels)
```

step 4

```python
train_ds = Dataset.from_pandas(train_df[["text", "label_id"]])
dev_ds = Dataset.from_pandas(dev_df[["text", "label_id"]])
test_ds = Dataset.from_pandas(test_df[["text", "label_id"]])
```

```python
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def preprocess(batch):
    return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=512)

train_ds = train_ds.map(preprocess, batched=True)
dev_ds = dev_ds.map(preprocess, batched=True)
test_ds = test_ds.map(preprocess, batched=True)

train_ds = train_ds.rename_column("label_id", "labels")
dev_ds = dev_ds.rename_column("label_id", "labels")
test_ds = test_ds.rename_column("label_id", "labels")

train_ds.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
dev_ds.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
test_ds.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

step 6: Load model for finetuning

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./legalbert-finetuned",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=100,
    do_train=True,
    do_eval=True,
    save_total_limit=2,
    fp16=False
)

def compute_metrics(p):
    preds = np.argmax(p.predictions, axis=1)
    labels = p.label_ids
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds, average="weighted")
    }

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=dev_ds,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

Explainability

```python
# ✅ Load sentence embedding model
sbert_model = SentenceTransformer("all-MiniLM-L6-v2")

# Embed all statutes
statutes["embedding"] = statutes["text"].apply(lambda x: sbert_model.encode(x, convert_to_tensor=True))
```

```python
def explain_case(case_text, top_k=3):
    # Split case text into sentences
    sentences = [s for s in case_text.split('.') if len(s.strip()) > 10]
    sent_embeds = sbert_model.encode(sentences, convert_to_tensor=True)

    results = []
    for i, row in statutes.iterrows():
        sim = util.cos_sim(sent_embeds, row["embedding"])
        max_sim = float(sim.max())
        best_sent = sentences[int(sim.argmax())]
        results.append({
            "statute_id": row["id"],
            "similarity": max_sim,
            "evidence_sentence": best_sent
        })

    ranked = sorted(results, key=lambda x: x["similarity"], reverse=True)[:top_k]
    return pd.DataFrame(ranked)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred

    # Apply sigmoid since multi-label uses independent probabilities per class
    probs = 1 / (1 + np.exp(-logits))

    # Convert probabilities to binary (multi-hot) predictions
    preds = (probs >= 0.5).astype(int)

    # Compute metrics
    f1 = f1_score(labels, preds, average="weighted", zero_division=0)
    precision = precision_score(labels, preds, average="weighted", zero_division=0)
    recall = recall_score(labels, preds, average="weighted", zero_division=0)

    # Exact match accuracy (all labels correct)
    acc = (preds == labels).all(axis=1).mean()

    return {"accuracy": acc, "precision": precision, "recall": recall, "f1": f1}
```

```python
# 0) (Optional) pip installs if needed
# !pip install -q transformers datasets scikit-learn sentence-transformers torch

# 1) Imports
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import f1_score, precision_score, recall_score
from datasets import Dataset
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments
)

# 2) Paths — change these to your actual files
TRAIN_PATH = "ilsi/train.parquet"
DEV_PATH   = "ilsi/dev.parquet"
TEST_PATH  = "ilsi/test.parquet"
# STATUTE_PATH if you need later for explainability
# STATUTE_PATH = "ilsi/statute.parquet"

# 3) Load parquet files (robust)
def safe_load_parquet(path):
    df = pd.read_parquet(path)
    # Ensure expected columns exist: 'text' and label column (could be 'labels' or 'id' etc.)
    # Adjust below if your label column name differs.
    return df

train_df = safe_load_parquet(TRAIN_PATH)
dev_df   = safe_load_parquet(DEV_PATH)
test_df  = safe_load_parquet(TEST_PATH)

# 4) Normalize columns: make sure 'text' contains string and 'labels' contains list-like
def normalize_text_col(df, text_col="text"):
    df[text_col] = df[text_col].apply(lambda x: "" if x is None else x)
    # If stored as list tokens or np.array, join into a string
    def to_str(x):
        if isinstance(x, (list, tuple, np.ndarray)):
            return " ".join(str(i) for i in x)
        return str(x)
    df[text_col] = df[text_col].apply(to_str)
    return df

def safe_parse_labels_col(x):
    # Accept list / ndarray / string like "['Section 2', 'Section 5']" / single string
    if isinstance(x, list):
        return x
    if isinstance(x, (np.ndarray, tuple)):
        return list(x)
    if pd.isna(x):
        return []
    if isinstance(x, str):
        x = x.strip()
        # If it looks like a python list string, try ast.literal_eval
        if x.startswith("[") and x.endswith("]"):
            try:
                import ast
                val = ast.literal_eval(x)
                return list(val) if isinstance(val, (list, tuple, np.ndarray)) else [str(val)]
            except Exception:
                # fallback: split by comma
                cleaned = x.strip("[]")
                parts = [p.strip().strip("'\"") for p in cleaned.split(",") if p.strip()!=""]
                return parts
        # If comma-separated without brackets
        if "," in x:
            parts = [p.strip().strip("'\"") for p in x.split(",") if p.strip()!=""]
            return parts
        # otherwise single label string
        return [x]
    # fallback
    return []

# normalize
train_df = normalize_text_col(train_df, text_col="text")
dev_df   = normalize_text_col(dev_df, text_col="text")
test_df  = normalize_text_col(test_df, text_col="text")

# find label column automatically if not 'labels'
possible_label_cols = ["labels", "label", "id", "sections"]
label_col = None
for c in possible_label_cols:
    if c in train_df.columns:
        label_col = c
        break
if label_col is None:
    # fallback: try to find a column with list-like first element
    for c in train_df.columns:
        if any(isinstance(v, (list, tuple, np.ndarray)) for v in train_df[c].iloc[:10]):
            label_col = c
            break

if label_col is None:
    raise ValueError("Could not find label column automatically. Please set label_col to the name of your label column.")

# parse labels
train_df["labels_parsed"] = train_df[label_col].apply(safe_parse_labels_col)
dev_df["labels_parsed"]   = dev_df[label_col].apply(safe_parse_labels_col)
test_df["labels_parsed"]  = test_df[label_col].apply(safe_parse_labels_col)

# 5) MultiLabelBinarizer
mlb = MultiLabelBinarizer()
y_train = mlb.fit_transform(train_df["labels_parsed"])
y_dev   = mlb.transform(dev_df["labels_parsed"])
y_test  = mlb.transform(test_df["labels_parsed"])

num_labels = len(mlb.classes_)
print("Number of statute labels (columns):", num_labels)

# Add vector column
train_df["labels_vec"] = y_train.tolist()
dev_df["labels_vec"]   = y_dev.tolist()
test_df["labels_vec"]  = y_test.tolist()

# 6) Create HuggingFace Datasets
hf_train = Dataset.from_pandas(train_df[["text", "labels_vec"]].rename(columns={"labels_vec":"labels"}))
hf_dev   = Dataset.from_pandas(dev_df[["text", "labels_vec"]].rename(columns={"labels_vec":"labels"}))
hf_test  = Dataset.from_pandas(test_df[["text", "labels_vec"]].rename(columns={"labels_vec":"labels"}))

# 7) Tokenizer & Model
MODEL_NAME = "bert-base-uncased"  # or "nlpaueb/legal-bert-base-uncased" if available locally
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)

def tokenize_batch(batch):
    texts = [str(t) for t in batch["text"]]
    enc = tokenizer(texts, padding="max_length", truncation=True, max_length=256)
    enc["labels"] = batch["labels"]
    return enc

# 8) Tokenize (batched)
tokenized_train = hf_train.map(tokenize_batch, batched=True, remove_columns=["text"])
tokenized_dev   = hf_dev.map(tokenize_batch, batched=True, remove_columns=["text"])
tokenized_test  = hf_test.map(tokenize_batch, batched=True, remove_columns=["text"])

# 9) Format for PyTorch
tokenized_train.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_dev.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_test.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])

# 10) Build model for multi-label classification
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    problem_type="multi_label_classification"
)

# 11) Metrics for multi-label
from sklearn.metrics import precision_score, recall_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    # logits: (N, num_labels), labels: (N, num_labels)
    probs = 1 / (1 + np.exp(-logits))          # sigmoid
    preds = (probs >= 0.5).astype(int)        # threshold 0.5

    labels = labels.astype(int)

    # micro/macro/samples F1
    micro_f1 = f1_score(labels, preds, average="micro", zero_division=0)
    macro_f1 = f1_score(labels, preds, average="macro", zero_division=0)
    samples_f1 = f1_score(labels, preds, average="samples", zero_division=0)

    prec_micro = precision_score(labels, preds, average="micro", zero_division=0)
    rec_micro  = recall_score(labels, preds, average="micro", zero_division=0)

    # exact-match / subset accuracy: all labels must match
    exact_match = np.all(preds == labels, axis=1).mean()

    return {
        "micro_f1": float(micro_f1),
        "macro_f1": float(macro_f1),
        "samples_f1": float(samples_f1),
        "precision_micro": float(prec_micro),
        "recall_micro": float(rec_micro),
        "exact_match": float(exact_match)
    }

# 12) TrainingArguments - compatible with older transformers
training_args = TrainingArguments(
    output_dir="./bert_multilabel_out",
    num_train_epochs=1,                 # we'll loop manually for multiple epochs to allow manual eval
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=200,
    save_total_limit=2,
    do_train=True,
    do_eval=True,
    fp16=False
)

# 13) Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_dev,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

# 14) Manual per-epoch loop (so we can evaluate & save best manually)
NUM_EPOCHS = 3
best_metric = -1.0
best_dir = None

for epoch in range(NUM_EPOCHS):
    print(f"\n===== Epoch {epoch+1}/{NUM_EPOCHS} =====")
    trainer.train()               # runs for training_args.num_train_epochs (which is 1 here)
    eval_metrics = trainer.evaluate(eval_dataset=tokenized_dev)
    print("Eval metrics:", eval_metrics)
    # metric key will be like 'eval_micro_f1' because compute_metrics returns micro_f1
    metric_key = None
    for k in eval_metrics.keys():
        if k.endswith("micro_f1"):
            metric_key = k
            break
    if metric_key is None:
        # fallback to any metric
        metric_key = list(eval_metrics.keys())[-1]
    cur_metric = eval_metrics.get(metric_key, -1.0)
    # save if improved
    if cur_metric > best_metric:
        best_metric = cur_metric
        best_dir = f"./best_model_epoch_{epoch+1}"
        trainer.save_model(best_dir)
        print(f"New best model saved to {best_dir} (metric {metric_key}={cur_metric})")

print("Training finished. Best metric:", best_metric, "saved at:", best_dir)

# 15) Final evaluation on test set (use the best saved model)
if best_dir is not None:
    model = AutoModelForSequenceClassification.from_pretrained(best_dir, local_files_only=True)
    trainer.model = model

test_pred = trainer.predict(test_dataset=tokenized_test)
# compute metrics on test
test_logits = test_pred.predictions
test_labels = test_pred.label_ids
# reuse compute_metrics (it expects tuple)
test_metrics = compute_metrics((test_logits, test_labels))
print("Test metrics:", test_metrics)

# 16) Example of decoding predictions for one sample (top-k labels)
probs = 1 / (1 + np.exp(-test_logits))
sample_idx = 0
topk = 5
top_inds = np.argsort(-probs[sample_idx])[:topk]
print("Top predicted statutes for sample 0:")
for i in top_inds:
    print(mlb.classes_[i], float(probs[sample_idx][i]))

# 17) Save ML model & tokenizer final
trainer.save_model("./final_multilabel_model")
tokenizer.save_pretrained("./final_multilabel_model")

# 18) Quick tips
#print("\nTIPS:")
#print("- If training is slow, reduce max_length to 128 and/ or use a smaller model (distilbert).")
#print("- You can tune per-class thresholds instead of 0.5 by using dev set probabilities to pick best threshold per label.")
#print("- For explainability: use a sentence-level SBERT to find which sentence in the case text has highest cosine to each predicted statute's definition.")
```

```python
# ================================================================
# STEP 6 — Metrics (multi-label)
# ================================================================
from sklearn.metrics import f1_score, accuracy_score
import numpy as np

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = (logits > 0).astype(int)  # because multi-label uses sigmoid
    f1 = f1_score(labels, preds, average="micro")
    acc = (preds == labels).mean()
    return {"accuracy": acc, "f1": f1}

# ================================================================
# STEP 7 — Training arguments (without deprecated params)
# ================================================================
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./legalbert-finetuned",
    num_train_epochs=4,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=100,
    save_total_limit=2,
    do_eval=True
)


# ================================================================
# STEP 8 — Trainer
# ================================================================
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_dev,
    compute_metrics=compute_metrics,
)

# ================================================================
# STEP 9 — Train!
# ================================================================
trainer.train()

# ================================================================
# STEP 10 — Evaluate
# ================================================================
trainer.evaluate()
```
